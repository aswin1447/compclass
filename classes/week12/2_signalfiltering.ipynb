### Quick recap of Fourier transforms

Let's say you have a sample with N points over time T.

```python
def plot_with_fft(t, signal):
    f = np.fft.rfft(signal)
    ft = np.fft.rfftfreq(len(t), t[1] - t[0])
    
    fig, axs = plt.subplots(1,2, figsize=(7,3))
    axs[0].plot(t,signal)
    axs[1].plot(ft,f.real)
    axs[1].plot(ft,f.imag)
    axs[1].plot(ft,np.abs(f), color='black')
    plt.show()
```

```python
import numpy as np
```

```python
%matplotlib notebook
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 1_000
T = 2*np.pi

t = np.linspace(0, T, N, endpoint=False)
```

```python
y = 1 / (1 - .9 * np.sin(t))
y += np.random.rand(N)*2-1
```

```python
plot_with_fft(t, y)
```

```python
f = np.fft.rfft(y)
ft = np.fft.rfftfreq(N, T/N)
```

```python
fig, ax = plt.subplots()
ax.plot(ft,np.abs(f))
ax.set_yscale('log')
plt.show()
```
